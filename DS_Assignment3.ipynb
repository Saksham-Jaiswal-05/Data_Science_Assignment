1. Write a Python program to draw line charts of the financial data of
Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['10-03-16','10-04-16','10-05-16','10-06-16','10-07-16'],
    'Open': [774.25,776.03,779.31,779,779.66],
    'High': [776.065002,778.71,782.07,780.47998,779.659973],
    'Low': [769.5,772.89,775.65,775.539978,770.75],
    'Close': [772.56,776.43,776.47,776.86,775.08]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')

plt.title("Alphabet Inc Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

4. Assuming the same CSV file, i.e., monthsales.csv, plot the line chart
with following customisations. Chart should have Month name on Xaxis instead of numbers.
Maker =”*”
Marker size=10
linestyle=”–“
Linewidth =3


```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "2018": [2500, 2200, 3000, 2800, 2750, 3500],
    "2019": [2600, 1800, 2500, 3000, 3200, 2000],
    "2020": [2450, 2000, 3000, 3100, 1950, 3550],
    "2021": [3250, 3200, 3800, 2800, 2900, 2300]
}

df = pd.DataFrame(data)

plt.plot(df["Month"], df["2018"], marker="*", markersize=10, linestyle="--", linewidth=3, label="2018")
plt.plot(df["Month"], df["2019"], marker="*", markersize=10, linestyle="--", linewidth=3, label="2019")
plt.plot(df["Month"], df["2020"], marker="*", markersize=10, linestyle="--", linewidth=3, label="2020")
plt.plot(df["Month"], df["2021"], marker="*", markersize=10, linestyle="--", linewidth=3, label="2021")

plt.title("Year-Wise Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

6. Visit data.gov.in , search for the following in “catalogs” option of the
website:
• Final population Totals, India and states
• State Wise literacy rate Download them and create a CSV file containing
population data and literacy rate of the respective state.
Also add a column Region to the CSV file that should contain the values
East, West, North and South.
Plot a scatter plot for each region where X axis should be population and
Y axis should be Literacy rate. Change the marker to a diamond and size
as the square root of the literacy rate. Group the data on the column region
and display a bar chart depicting average literacy rate for each region.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'State': ['Uttar Pradesh','Maharashtra','Bihar','West Bengal','Madhya Pradesh','Tamil Nadu','Rajasthan','Karnataka','Gujarat','Andhra Pradesh','Odisha','Telangana','Kerala','Jharkhand','Assam','Punjab','Chhattisgarh','Haryana','Delhi','Jammu and Kashmir','Uttarakhand','Himachal Pradesh','Tripura','Meghalaya','Manipur','Nagaland','Goa','Arunachal Pradesh','Mizoram','Sikkim'],
    'Population': [199812341,112374333,104099452,91276115,72626809,72147030,68548437,61095297,60439692,84580777,41974218,35003674,33406061,32988134,31205576,27743338,25545198,25351462,16787941,12541302,10086292,6864602,3673917,2966889,2570390,1978502,1458545,1383727,1097206,610577],
    'Literacy Rate': [67.68,82.34,61.80,76.26,69.32,80.09,66.11,75.36,78.03,67.02,72.87,66.54,94.00,66.41,72.19,75.84,70.28,75.55,86.21,68.74,78.82,82.80,87.75,74.43,79.85,79.55,88.70,65.38,91.33,81.42],
    'Region': ['North','West','East','East','Central','South','North','South','West','South','East','South','South','East','East','North','Central','North','North','North','North','North','East','East','East','East','West','East','East','East']
}
df = pd.DataFrame(data)

for r in df['Region'].unique():
    d = df[df['Region'] == r]
    plt.scatter(d['Population'], d['Literacy Rate'], marker='D', s=np.sqrt(d['Literacy Rate'])*10)
    plt.title(f'{r} Region')
    plt.xlabel('Population')
    plt.ylabel('Literacy Rate')
    plt.show()

df.groupby('Region')['Literacy Rate'].mean().plot(kind='bar')
plt.title('Average Literacy Rate by Region')
plt.ylabel('Literacy Rate (%)')
plt.show()
```

8. Write a Python program to create bar plot of scores by group and gender.
Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
The code snippet gives the output shown in the following screenshot:Screenshot 2026-04-05 185348.png

```python
import numpy as np
import matplotlib.pyplot as plt

men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

x = np.arange(len(men))
width = 0.35

plt.bar(x - width/2, men, width, label='Men')
plt.bar(x + width/2, women, width, label='Women')

plt.xlabel("Groups")
plt.ylabel("Scores")
plt.title("Scores by Gender")
plt.xticks(x, ['G1','G2','G3','G4','G5'])
plt.legend()

plt.show()
```
