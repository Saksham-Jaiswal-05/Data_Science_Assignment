```python
# QUESTION 1 - Employee DataFrame

emp = pd.DataFrame({
    "Name": ["Atul","Raj","Darpan","Anmol","Piyush"],
    "Age": [25,30,23,22,32],
    "Designation": ["Manager","Clerk","Analyst","Clerk","Manager"],
    "Salary": [56000,25000,35000,28000,58000],
    "Bonus": [15000,7000,9000,13000,12000]
}, index=["E101","E102","E103","E104","E105"])


emp["City"] = None
emp["Shift"] = None

emp.loc["E106"] = ["GEET", 24, "Analyst", 70000, 15000, "Tilak Nagar", "Morning"]


emp["Shift"] = "Morning"


print("\nYoungest Employee:")
print(emp[emp["Age"] == emp["Age"].min()])


print("\nAll Clerks:")
print(emp[emp["Designation"] == "Clerk"])


print("\nAverage Bonus:")
print(emp["Bonus"].mean())



print("\nTotal Salary:")
print(emp["Salary"].sum())


emp.loc["E101":"E105","City"] = [
    "Janak Puri",
    "Tilak Nagar",
    "Shiv Nagar",
    "Vikas Puri",
    "Janak Puri"
]


emp = emp.drop(["E103","E105"])


emp = emp.rename(columns={"Bonus": "Incentive"})


print("\nFinal DataFrame:")
print(emp)
```

```python
# QUESTION 2: DataFrame “data”

import pandas as pd

data = pd.DataFrame({
    "Count":[15,20,25,35],
    "Price":[250,200,600,900]
}, index=["Pencil","Pen","Eraser","Scale"])

print(data)

data.loc["Eraser","Count"] = 15

data = data.rename(index={"Scale":"Wooden Scale"})

data.index = ["Product 1","Product 2","Product 3","Product 4"]

data.columns = ["Discount","Total"]

data.loc["Product 5"] = [10,100]

Products = data.drop(columns=["Total"])

print(data)
print(Products)
```

```python
# QUESTION 3: Production DataFrame

import pandas as pd
import numpy as np

prod = pd.DataFrame({
    "Fruits":[44,11950,7152,140169,7830,113],
    "Pulses":[23,818,33,3184,931,1],
    "Rice":[814,1930,11586,13574,7452,2604],
    "Wheat":[5,2737,16440,30056,np.nan,np.nan]
}, index=["Tripura","Gujrat","Punjab","UP","AP","Kerala"])

print(prod)

print(prod.sum(axis=1))

print(prod.sum(axis=0))

print(prod["Wheat"].count())
```
